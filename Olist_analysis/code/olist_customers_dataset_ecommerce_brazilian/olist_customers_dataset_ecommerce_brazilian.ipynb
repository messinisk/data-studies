```python
import pandas as pd
import matplotlib.pyplot as plt
from caas_jupyter_tools import display_dataframe_to_user

df = pd.read_csv("/mnt/data/olist_customers_dataset_ecommerce_brazilian.csv")

# State distribution
state_counts = df["customer_state"].value_counts().rename_axis("state").reset_index(name="customers")
state_counts["share_pct"] = (state_counts["customers"] / len(df) * 100).round(2)

# Top cities overall
city_counts = df["customer_city"].value_counts().rename_axis("city").reset_index(name="customers")
city_counts["share_pct"] = (city_counts["customers"] / len(df) * 100).round(2)

# Cities within top 5 states
top_states = state_counts["state"].head(5).tolist()
city_top_by_state = (
    df[df["customer_state"].isin(top_states)]
    .groupby(["customer_state", "customer_city"])
    .size()
    .reset_index(name="customers")
    .sort_values(["customer_state", "customers"], ascending=[True, False])
)

# Prepare summary frames
top10_states = state_counts.head(10)
top20_cities = city_counts.head(20)

display_dataframe_to_user("Top 10 states by customer records", top10_states)
display_dataframe_to_user("Top 20 cities by customer records", top20_cities)

# Plot: state distribution (all states)
plt.figure()
plt.bar(state_counts["state"], state_counts["customers"])
plt.title("Customer records by Brazilian state (UF)")
plt.xlabel("State (UF)")
plt.ylabel("Customer records")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Plot: top 20 cities
plt.figure()
plt.bar(top20_cities["city"], top20_cities["customers"])
plt.title("Top 20 customer cities")
plt.xlabel("City")
plt.ylabel("Customer records")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Show top 5 cities for each of top 5 states
top5_cities_each_top_state = (
    city_top_by_state.groupby("customer_state").head(5).reset_index(drop=True)
)
display_dataframe_to_user("Top 5 cities within each of the top 5 states", top5_cities_each_top_state)

```
